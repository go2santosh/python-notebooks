# Learning Celcious to Fahrenheit using Keras

```python
import tensorflow as tf
import numpy as np
```

```python
# training data
x = np.array([-50, -40, -30, -20, -10, 0, 10, 20, 30, 40, 50])
y = np.array([-58, -40, -22, -4, 14, 32, 50, 68, 86, 104, 122])
```

```python
# build the model
model = tf.keras.models.Sequential()
model.add(tf.keras.Input(shape=(1,)))
model.add(tf.keras.layers.Dense(1, activation=None))
```

```python
# compile model
model.compile(optimizer=tf.keras.optimizers.Adam(0.1), loss=tf.keras.losses.mean_squared_error)
```

```python
# train the model
history = model.fit(x=x, y=y, epochs=1000)
```

```python
# training information
import matplotlib.pyplot as plt
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.plot(history.history['loss'])
```

```python
# predict test values
print(model.predict(x=[36]))
```

```python
# learned weights
print(model.get_weights())
```
